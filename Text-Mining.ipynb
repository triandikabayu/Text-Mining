```python
stemer = PorterStemmer()
```

```python
df = pd.read_csv('/content/drive/MyDrive/imdbkaggle/imdb_labelled.csv', header=None)
df.head()
```

```python
# Membersihkan data
def clean_text(text):
    # Hapus karakter yang tidak diinginkan
    text = re.sub('[^a-zA-Z]', ' ', text)

    # Konversi menjadi lowercase
    text = text.lower()

    # Tokenisasi teks menjadi kata-kata
    words = word_tokenize(text)

    # Hapus stopwords
    words = [word for word in words if word not in stopwords.words('english')]

    # Stemming kata-kata
    words = [stemer.stem(word) for word in words]

    # Gabungkan kembali kata-kata menjadi teks
    text = ' '.join(words)

    return text

# Terapkan fungsi clean_text pada kolom teks (kolom 0)
df['clean_text'] = df[0].apply(clean_text)

# Tampilkan hasil setelah pembersihan terdapat pada kolom akhir (clean_text)
df.head()
```
